```


import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM, GRU
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
from keras.layers import MaxPooling1D
from keras.layers import Flatten, GlobalMaxPooling1D, GlobalAveragePooling1D
from keras.layers import Dropout, BatchNormalization, Input, SpatialDropout1D
from keras.layers import Dense, Bidirectional, concatenate
from keras.models import Model
from keras.optimizers import Adam
from keras_self_attention import SeqSelfAttention
import keras.backend as K
from keras.models import load_model
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.layers import Dense,Input
from numpy import asarray
t  = Tokenizer(num_words=40000)

model_dwn = load_model('/content/drive/My Drive/unbiased_project/unbiased_model.hdf5',custom_objects={'SeqSelfAttention': SeqSelfAttention})

x=[]
f = open('/content/drive/My Drive/unbiased_project/train_x.txt', 'r')
x = f.readlines()
f.close()

t.fit_on_texts(x)

```

```

import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
```

```
stop_words = ['a', 'and']
```

```
sample_texts = "i am a lesbian, i dont care what other speak about me . i am in love with my life. i had rape experience, but i survived "
text_tokens = word_tokenize(sample_texts)

tokens_without_sw = [word for word in text_tokens if not word in stop_words]
# sample_txt_vect= vectorizer.transform(sample_txt)
# pred_lg=lgmodel.predict(sample_txt_vect) 
# print(pred_lg)
filtered_sentence = (" ").join(tokens_without_sw)
print(filtered_sentence)
encoded_sample = t.texts_to_sequences([filtered_sentence])
# defining a max size for padding.
max_len = 150
# padding the vectors of each datapoint to fixed length of 600.
pad_sample = pad_sequences(encoded_sample,maxlen = max_len,padding='post')
preds = model_dwn.predict(pad_sample)
print(preds)
```
